###  How many values are we missing?

```python
print('Missing vals:\n\n')
for attr in car_info: #for each attribute of car_info 
    c = car_info[car_info[attr] == '?'] #select all missing val instances for attribute
    if not c.empty: #if there are missing values for attribute
        print(attr, ': ', len(c)) #print number of missing values for attribute
```

### Are some numerical variables missing in the above grid?
It appears some variables which should be numeric not are showing in car_info.describe() <br>
Let's check variable types

```python
car_info.dtypes
```

We see that some variables which should be numeric are instead of type `object`, <br>
<br> <br>
Let's fix that:

```python
list_of_variables_of_type_obj = ['normalized_losses', 'bore', 
                                'stroke', 'compression_ratio', 
                                'horsepower', 'peak_rpm', 'price'
                                ]
car_info_2 = car_info
#replace '?', a string, with NaN to allow for numpy operations
car_info.replace('?', np.NaN, inplace=True)
for var in list_of_variables_of_type_obj:
    car_info[var] = pd.to_numeric(car_info[var])
#confirm that we now have more numerical values
car_info.dtypes
```
