```python
cities = ['Los Angeles', 'Phoenix', 'San Diego', 'San Francisco', 'Seattle']
columns = ['Unnamed: 0', 'Unnamed: 1']
#convert the original wide to long using melt method
long_flights = flights_df.melt(id_vars=columns, value_vars = cities)
new_cols = ['id', 'airline', 'status', 'city', 'count']
long_flights.rename(columns={'Unnamed: 0': 'airline', 'Unnamed: 1':'status', 'variable':'city', 'value':'count'}, inplace=True)
#fill na airlines with last seen airline
long_flights['airline'].fillna(method='ffill', inplace = True)
#remove observations with 'na' flight status
long_flights = long_flights[long_flights['status'].notna()]
long_flights
```
